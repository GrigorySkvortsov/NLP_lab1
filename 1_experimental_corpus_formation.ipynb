```python
import pandas as pd
import json
from tqdm import tqdm

input_file = "C:/Users/User/.cache/kagglehub/datasets/yutkin/corpus-of-russian-news-articles-from-lenta/versions/2/lenta-ru-news.csv"
output_file = "0_news_corpus.jsonl"

# –ó–∞–≥—Ä—É–∂–∞–µ–º CSV
df = pd.read_csv(input_file)

# –ü—Ä–∏–≤–æ–¥–∏–º –∏–º–µ–Ω–∞ –∫–æ–ª–æ–Ω–æ–∫ –∫ –µ–¥–∏–Ω–æ–æ–±—Ä–∞–∑–Ω–æ–º—É –≤–∏–¥—É
df.columns = [c.strip().lower() for c in df.columns]

# –ü—Ä–æ–≤–µ—Ä–∏–º, —á—Ç–æ –≤—Å–µ –Ω—É–∂–Ω—ã–µ –ø–æ–ª—è –µ—Å—Ç—å
expected_cols = {"url", "title", "text", "date", "topic", "tags"}
missing = expected_cols - set(df.columns)
if missing:
    print(f"–í–Ω–∏–º–∞–Ω–∏–µ: –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç –∫–æ–ª–æ–Ω–∫–∏ {missing}")

# –°–æ–∑–¥–∞—ë–º —Å–ø–∏—Å–æ–∫ —Å—Ç–∞—Ç–µ–π
records = []
for _, row in tqdm(df.iterrows(), total=len(df), desc="–ö–æ–Ω–≤–µ—Ä—Ç–∞—Ü–∏—è"):
    record = {
        "url": str(row.get("url", "")).strip(),
        "title": str(row.get("title", "")).strip(),
        "text": str(row.get("text", "")).strip(),
        "date": str(row.get("date", "")).strip(),
        "category": str(row.get("topic", "") or row.get("tags", "")).strip(),
    }
    # —Ñ–∏–ª—å—Ç—Ä—É–µ–º —Å–ª–∏—à–∫–æ–º –∫–æ—Ä–æ—Ç–∫–∏–µ —Ç–µ–∫—Å—Ç—ã
    if len(record["text"].split()) >= 30:
        records.append(record)

# –°–æ—Ö—Ä–∞–Ω—è–µ–º –≤ JSONL
with open(output_file, "w", encoding="utf-8") as f:
    for r in records:
        json.dump(r, f, ensure_ascii=False)
        f.write("\n")

print(f"–°–æ—Ö—Ä–∞–Ω–µ–Ω–æ {len(records)} –∑–∞–ø–∏—Å–µ–π –≤ {output_file}")
```
